```python
def AND(x1, x2):
    w1 , w2, theta =0.5 , 0.5 , 0.7
    temp = x1*w1 + x2*w2
    if temp <= theta:
        return 0;
    elif temp>theta:
        return 1;
```

```python
print AND(0,0)
print AND(1,0)
print AND(0,1)
print AND(1,1)
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    
    temp = np.sum(w*x)+b
    if temp <=0:
        return 0
    else :
        return 1
```

```python
def OR(x1 ,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = -0.2
    
    temp = np.sum(w*x)+b
    if temp <=0:
        return 0
    else :
        return 1
```

```python
print NAND(0,0)
print NAND(0,1)
print NAND(1,0)
print NAND(1,1)
```

```python
print OR(0,0)
print OR(1,0)
print OR(0,1)
print OR(1,1)
```

```python
def XOR(x1 ,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y  = AND(s1,s2)
    return y
```

```python
print XOR(0,0)
print XOR(1,0)
print XOR(0,1)
print XOR(1,1)
```

```python
def step(x):
    y=x>0
    y=y.astype(np.int)
    return y
```

```python
y=y.astype(np.int) :false ->0 , true -> 1 로 바꾸는 명령문
x=np.arange(-2.0, 2.0, 0.001)   : x축 최소값, x축 최대값, 간격 0.001
```

```python
x=np.arange(-2.0, 2.0, 0.001)
y=step(x)
plt.ylim(-2,2)
plt.plot(x,y)
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5 , 5 , 0.0001)
y=relu(x)
plt.plot(x,y)
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.title('Relu')
```
